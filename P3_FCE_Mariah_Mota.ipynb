## 1. Identificação do Problema

O sistema analisado é um wetland construído utilizado para o tratamento de soro de leite, um resíduo típico da indústria de laticínios. Dados experimentais, realizados durante os meses de março, abril e maio, coletados por Eduardo Neiva, para uso em Iniciação Científica, estãao sendo utilizados neste trabalho. O arquivo possui dados de diferentes parâmetros físico-químicos do efluente tratado, como, pH, turbidez, condutividade elétrica, sólidos suspensos (SS), sólidos dissolvidos (SD) e sólidos totais (ST).

A pergunta principal é:
Como esses parâmetros variam ao longo do tempo e o que indicam em relação à eficiência do sistema de tratamento?

Em outras palavras, busca-se avaliar se os wetlands construídos foram capazes de promover melhorias nas características da água residual, contribuindo para sua qualidade final.


---


## 2. Formulação do Problema

A análise será feita de forma exploratória e comparativa, com foco em:

Análise exploratória: verificar o comportamento temporal dos parâmetros (tendências, oscilações e padrões nos dados).

Comparação entre parâmetros: observar como cada variável se relaciona, por exemplo, se uma redução em turbidez acompanha a redução em sólidos suspensos.


Não foram utilizadas técnicas de classificação ou regressão, a ideia é a análise descritiva e visualização gráfica dos dados experimentais.

---


## 3. Coleta e Tratamento de Dados

Origem dos dados: experimental, obtido durante a iniciação científica de um dos autores.

Formato: planilha Excel (.xlsx) com registros diários dos parâmetros entre março e maio.

Tratamento dos dados:

Remoção de colunas irrelevantes (“Unnamed”) que vieram do Excel.

Exclusão de linhas vazias.

Padronização dos índices para facilitar a manipulação.

Para a visualização, serão construídos gráficos de séries temporais (valores ao longo dos dias) para cada parâmetro, permitindo identificar tendências e comportamentos anormais.

---


## 4. Implementação Computacional

**BIBLIOTECAS UTILIZADAS**

```python
import pandas as pd
import matplotlib.pyplot as plt
```

**COLETA DE DADOS DO ARQUIVO**

```python
df = pd.read_excel('dados_ic_FCE.xlsx')
df.head()
```

**LIMPEZA DOS DADOS**

```python
df.drop(['Unnamed: 0', 'Unnamed: 1', 'Unnamed: 2', 'Unnamed: 4'], axis=1, inplace=True)
df.dropna(inplace=True)
df.reset_index(drop=True, inplace=True)
df
```

**Visualização gráfica**

```python
plt.figure(figsize=(8, 5))
plt.plot(df['Dia'], df['CONDUTIVIDADE'], label='Condutividade (μS/cm)', marker='o')
plt.plot(df['Dia'], df['TURBIDEZ'], label='Turbidez (NTU)', marker='o')
plt.plot(df['Dia'], df['SD'], label='Sólidos dissolvidos (ppm)', marker='o')
plt.plot(df['Dia'], df['SS'], label='Sólidos suspensos (ppm)', marker='o')
plt.plot(df['Dia'], df['ST'], label='Sólidos totais (ppm)', marker='o')
plt.xlabel('Dia')
plt.ylabel('Valores')
plt.xticks(rotation=45)   # gira os rótulos em 45 graus
plt.legend()
plt.show()

plt.figure(figsize=(5, 5))
df.plot(x='Dia', y='PH', marker='o', legend=False)
plt.xlabel('Dia')
plt.ylabel('pH')
plt.xticks(rotation=45)
plt.show()
```

## 5. Validação e Verificação




Os resultados obtidos a partir da análise gráfica mostram-se consistentes com o comportamento esperado para wetlands construídos tratando soro de leite.

  - **Condutividade** permaneceu elevada, o que era esperado, pois wetlands não removem sais dissolvidos com eficiência.

- **pH (gráfico 2):**

  - Tendência de aumento em abril e maio, aproximando-se da neutralidade.
  - Comportamento coerente com processos biológicos acontecem dentro do tanque.

- **Consistência geral:**
  - Tendências seguem o esperado para tratamento biológico: redução parcial de turbidez e sólidos, manutenção da condutividade e ajuste gradual do pH.
  - Os gráficos são coerentes e compatíveis com a literatura.
